## Descriptive Analytics for Numerical Columns

### Load the dataset into a data analysis tool or programming environment (e.g., Python with pandas library).

```python
#import pandas library
import pandas as pd
#read csv file
df=pd.read_csv("sales_data_with_discounts.csv")
df
```

### Identify numerical columns in the dataset.

```python
#Fetch and display the numerical columns from dataset
df.select_dtypes(include='number')
```

### Calculate the mean, median, mode, and standard deviation for these columns.

##### Mean

```python
#Calculate mean of all numerical columns
mean_value=df.select_dtypes(include='number').mean()
display("----- Mean_Value: -----",mean_value)
```

##### Median

```python
#Calculate median of all numerical columns
median_value=df.select_dtypes(include='number').median()
display("----- Median_Value: -----",median_value)
```

##### Mode

```python
#Calculate mode of all numerical columns
mode_value=df.select_dtypes(include='number').mode()
display("----- Mode_Value: -----",mode_value)
```

```python
#Calculate standard deviation of all numerical columns
std_value=df.select_dtypes(include='number').std()
display("----- Satandard deviation: -----",std_value)
```

#### Provide a brief interpretation of these statistics.

##### - Mean: 
The average value of a numeric column. It is calculated by adding all the values in column and dividing by the total number of values.        
##### - Median: 
The middle value in a dataset when the data is sorted in ascending order. If there is an even number of values, the median is the average of the two middle values. 

##### - Mode:
The values that appear most frequently in the dataset.A dataset can have one mode, more than one mode, or no mode.

##### - Standard deviation:
A measure of how spread out the values are from the mean.

## Data Visualization

#### Plot histograms for each numerical column.

```python
#import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Select numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

#plot histogram for each numerical columns one by one and label the x and y axis
for colms in numeric_columns:
    sns.histplot(data=df, x=colms)
    plt.title(f'Distribution of {colms}')
    plt.xlabel(colms)
    plt.ylabel('Range')
    plt.show()
```

#### 	Analyze the distribution (e.g., skewness, presence of outliers) and provide inferences.

```python
# Select numeric columns
numeric_cols = df.select_dtypes(include='number').columns

for cols in numeric_cols:
    #find Skewness for numerical columns
    cols_skew=df[cols].skew()    

    #find Kurtosis for numerical columns
    kurt_cols=df[cols].kurtosis()

    #Description for numerical columns
    describe_cols=df[cols].describe()
 
    #find Outliers 
    Q1 = df[cols].quantile(0.25)
    Q3 = df[cols].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers=df[cols][(df[cols] < lower_bound) | (df[cols] > upper_bound)]

    print(cols)
    print(f"Skewness of {cols}: ",cols_skew)
    print(f"Kurtosis of {cols}: ",kurt_cols)    
    print(f"Outliers of {cols}: ",len(outliers))
    print(f"**Description:**\n{cols}",describe_cols)
    print("\n")

```

#### inferences
- Skewness: Volume', 'Avg Price', 'Total Sales Value', 'Discount Amount', and 'Net Sales Value' all show positive skewness, with skewness values greater than 1, indicating that distribution is positively skewed (right-skewed),
 'Discount Rate (%)' shows negative skewness with a value of -1.0623, indicating an asymmetry with a tail extending towards lower values.

- Kurtosis: For Volume', 'Avg Price', 'Total Sales Value', 'Discount Amount', and 'Net Sales Value' distribution has heavier tails and sharper peak than a normal distribution (leptokurtic).
 For 'Discount Rate (%)' distribution is similar to a normal distribution (mesokurtic)

- Outliers: Each columns have outliers


### Boxplot

#### 	Create boxplots for numerical variables to identify outliers and the interquartile range.

```python
# Select numeric columns from dataset
numeric_columns = df.select_dtypes(include=['number']).columns

#plot boxplot for each numerical columns.
for colms in numeric_columns:
    sns.boxplot(data=df, y=colms)
    plt.title(f'Boxplot of {colms}')
    plt.ylabel(colms)
    plt.show()
```

### Discuss any findings, such as extreme values or unusual distributions.

```python
# Select numeric columns from dataset
numeric_cols = df.select_dtypes(include=['number']).columns


for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(col)
    print(f"Interquartile Range (IQR): {IQR}")
    print(f"Lower Bound (Q1 - 1.5*IQR): {lower_bound}")
    print(f"Upper Bound (Q3 + 1.5*IQR): {upper_bound}")
    print(f"Number of outliers: {len(outliers)}\n")
```

##### 
- Volume, Avg Price, Total Sales Value, Discount Amount, and Net Sales Value: The boxplots for these columns clearly show a significant number of data points extending in the upper limit. This shows the positive skewness.
##### 
- Discount Rate (%): The boxplot for 'Discount Rate (%)' shows data points extending below the lower limit. This shows the negative skewness.

## Standardization of Numerical Variables

#### 	Explain the concept of standardization (z-score normalization).

#### Explanation:
Standardization, or z-score normalization, is a data preprocessing technique used to scale numerical features so that they have a mean of 0 and a standard deviation of 1, making it more suitable for various machine learning algorithms and improving the overall analysis process.

#### Standardize the numerical columns using the formula: z=x-mu/sigma
#### Show before and after comparisons of the data distributions.

```python
# Select numeric columns
numeric_columns = df.select_dtypes(include=['number'])

mean_value=df.select_dtypes(include='number').mean()
std_value=df.select_dtypes(include='number').std()

standardized_data=(numeric_columns-mean_value)/std_value    
display("*********** Data before Standardization: *****************",numeric_columns,"*********** Data after Standardization: **********",standardized_data)
```

## Conversion of Categorical Data into Dummy Variables

#### 	Discuss the need for converting categorical data into dummy variables (one-hot encoding).


##### Explanation:
One-hot encoding is a technique used to convert categorical variables into a numerical format that can be used by machine learning for processing the dataset and further analysis.One-hot encoding creates new binary columns for each unique category in a categorical variable, where a '1' indicates the presence of that category and a '0' indicates that absence of the category. 

#### Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.
#### Display a portion of the transformed dataset.

```python
# Get categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

# Apply one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_columns,dtype=int)

#Display all columns
pd.set_option('display.max_columns', None)

# Display a portion of the transformed dataset
df_encoded.head(20)
```
